```python
import os
import time

# Path to the file being downloaded
file_path = '/teamspace/studios/this_studio/Sam_LoRA/sam_lora_model_train_250_freetown_haiti_eastafrica.pth'

# Expected size of the file in bytes (replace with the actual expected size)
expected_size = 10485760  # For example, 10 MB (10 * 1024 * 1024)

# Function to get the current size of the file
def get_file_size(path):
    return os.path.getsize(path) if os.path.exists(path) else 0

# Periodically check the file size
while True:
    current_size = get_file_size(file_path)
    print(f"Current size of the file '{file_path}': {current_size} bytes")

    if current_size >= expected_size:
        print("File has finished downloading.")
        break
    
    # Wait for a while before checking again
    time.sleep(5)  # Check every 5 seconds

print("Done.")
```
