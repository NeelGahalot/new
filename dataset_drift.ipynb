```python
import pandas as pd

# Read the CSV file into a DataFrame  freetown_deeplab_confidence_processed.csv
df = pd.read_csv('/teamspace/studios/this_studio/training_random_deeplab_confidence_processed.csv')

training_list = df['Deeplab Confidence'].tolist()
```

```python
df = pd.read_csv('/teamspace/studios/this_studio/production_random_deeplab_confidence_processed.csv')

production_list = df['Deeplab Confidence'].tolist()
```

```python
print(len(training_list))
```

```python

print(len(training_list))
print(training_list[:10])
# Function to check if a string can be converted to a float
def is_number(s):
    try:
        float(s)
        return True
    except ValueError:
        return False

# Convert elements to float, filtering out non-numeric strings
training_floats = [float(i) for i in training_list if is_number(i) and float(i) != 0.0]


```

```python

print(len(production_list))
print(production_list[:10])
# Function to check if a string can be converted to a float
def is_number(s):
    try:
        float(s)
        return True
    except ValueError:
        return False

# Convert elements to float, filtering out non-numeric strings
production_floats = [float(i) for i in production_list if is_number(i) and float(i) != 0.0]

print(len(production_floats))
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import ks_2samp
```

```python
plt.figure(figsize=(12, 6))
sns.kdeplot(production_floats, shade=True, label='Production')
sns.kdeplot(training_floats, shade=True, label='Training')
plt.legend(loc='upper right')
plt.title('KDE Plot')
plt.show()
```

```python
ks_stat, ks_p_value = ks_2samp(production_floats, training_floats)
print(f"KS Test Statistic: {ks_stat}")
print(f"KS Test P-value: {ks_p_value}")

if ks_p_value < 0.05:
    print("The distributions are significantly different.")
else:
    print("The distributions are not significantly different.")
```
