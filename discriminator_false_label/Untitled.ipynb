```python
from PIL import Image
import matplotlib.pyplot as plt
import os
def show(filename, dataset_dir = '/teamspace/studios/this_studio/discriminator_false_label/dataset'):
    img = Image.open(os.path.join(dataset_dir,'samples',filename) + '.jpg')
    plt.imshow(img)
    plt.show()
    mask = Image.open(os.path.join(dataset_dir,'binary_masks',filename) + '_binarymask.jpg')
    plt.imshow(mask)
    plt.show()
    label_path = os.path.join(dataset_dir,'labels',filename) + '_label.txt'
    with open(label_path, 'r') as f:
        print(f.readline())
    

```
